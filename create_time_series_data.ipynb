```python
g4_df = dt.fread('dexcomG4.csv').to_pandas()
g4_index = g4_df['MDR_REPORT_KEY'].copy()
#g4_index.to_csv('Index/g4index.csv')
```

```python
def create_ts_df(df, freq):
    df['DATE'] = pd.to_datetime(df['DATE_RECEIVED'], format = '%Y-%m-%d')
    df = df.set_index('DATE')
    df['tot'] = 1
    df = df.resample(freq).sum()
    df_tot = df.groupby(df.index)['tot'].agg('sum').to_frame()
    return df_tot
```

```python
FREQ = 'M'
```

```python
g6_tot = create_ts_df(g6_df, freq = FREQ)
g5_tot = create_ts_df(g5_df, freq = FREQ)
g4_tot = create_ts_df(g4_df, freq = FREQ)
```

```python
data_frames = {'G4': g4_tot, 'G5': g5_tot, 'G6': g6_tot}
```

```python
for device in data_frames:
    plt.plot(data_frames.get(device), label = 'Plot for {0}'.format(device))

plt.legend()
plt.show()
```
